```python
#import beautifulsoup and urllib.request classes

from bs4 import BeautifulSoup as bs
import urllib.request as ulib
import pprint as pp
import re
from datetime import datetime as dt

class QuestionSet:
    #bs_data stands for beautifulsoup
    #difficulty is not being used currently
    def __init__(self, url, mirrors=[], difficulty=None, bs_data=None):
        self.dburl = url
        self.mirrors = mirrors
        self.difficulty = difficulty
        self.bs_data = bs_data
    
    #adds mirror to list of mirrors
    def add_mirror(self, mirror):
        self.mirrors.append(mirror)
    
    #opens up url for questionset and parses data into a beautifulsoup object
    def get_dbdata(self):
        self.bs_data = bs(ulib.urlopen(self.dburl).read())

class Mirror:
    
    def __init__(self, url, title=None, date=None, loc=None, addr=None, bs_data=None):
        self.url, self.title, self.date = url, title, date
        self.loc, self.addr, self.bs_data = loc, addr, bs_data

    def __repr__(self):
        ret_str = "Mirror name: " + self.title + "\n"
        ret_str += "Mirror date: " + self.date.strftime("%B %d,%Y")+"\n"
        ret_str += "Mirror location: " + self.loc + "\n"
        ret_str += "Mirror address: " + self.addr + "\n"
        return ret_str
    
    def set_title_and_desc(self):
        '''
        <DIV class="MultilineHeading">
            <H2>ACF Fall 2014: HS mirror</H2>
            <H5>High school tournament on November 08, 2014</H5>
        </DIV>
        '''
        data = self.bs_data.findAll("div", {'class':'MultilineHeading'})
        self.title = parse_line(data[0].contents[1])[2] #I don't know of more elegant ways to parse this yet
        self.tag = parse_line(data[0].contents[3])[2]
        return
   
    def set_loc_and_addr(self):
        '''
        Right the location name and address are being parsed here.
                            <P><SPAN class="FieldName">Host location:</SPAN> NIU</P>
                            <P><SPAN class="FieldName">Address:</SPAN> DuSable Hall, DeKalb, Illinois 60115</P>
                    
        '''
        for datum in self.bs_data.findAll('span', {'class':'FieldName'}):
            if 'Host' in datum.contents[0]:
                self.loc = str(datum.next_sibling)
                continue
            elif 'Address' in datum.contents[0]:
                self.addr = str(datum.next_sibling)
                continue
        #it's possible that there are some entries with no values. Handle that.
        if self.loc == None :
            self.loc = "Unknown"
        if self.addr == None :
            self.addr = "Unknown"
        return
    
    def set_date_from_tag(self):
        '''
        The only place to find the date for the
        tournament is from the end of the tagline.
        There ought to be a discrete field or a tag around it.
        Again, probably talk to Dan/Jeff about it.
        '''
        vals = self.tag.split(" ")
        self.date=dt.strptime((vals[-3] + vals[-2] + vals[-1]), "%B%d,%Y")
        return
    
    def get_dbdata(self):
        '''
        populates the relevant data for the mirror object
        calles helper functions in mirror class
        '''
        self.bs_data = bs(ulib.urlopen(self.url).read())
        self.set_title_and_desc() #get title and description
        self.set_loc_and_addr() #get location and address
        if self.tag != "":
            self.set_date_from_tag() #parse out the date from the description
        else :
            date = "Unknown"
        return

#setup globals
DB_URL = "http://www.hsquizbowl.org/db/"

#setting up some regex for making my life easier
delimiters = '>', '"', '<',
regexPattern = '|'.join(map(re.escape, delimiters))


#I am not sure if this function is in the right place 
#(it is called by main and the mirror class)
#but it works right now so I am not moving it.
def parse_line(text):
    '''Right now this is accomplishing the task of parsing out 
    the delims listed above and splitting up the string being
    passed in.
    
    This data coming in here is very different for different cases
    so this is generic right now.'''
    return re.split(regexPattern, str(text))


'''
Program starts here

Everything above is a helper function
'''
def main(url_string):
    
    global DB_URL
    
    qset = QuestionSet(url_string) #creates a QuestionSet object
    qset.get_dbdata() #gets data for the questionset
    
    '''
    We are interested in the "Tournaments" linked to our question set.
    They are listed as:
    <span class="Name"><a href="tournaments/2310/">ACF Fall 2014: HS mirror</a></span>
    ... etc.
    '''
    
    mirrors_raw=qset.bs_data.findAll('span',{'class':'Name'})
    for mirror in mirrors_raw:
        mirror_url = DB_URL + parse_line(mirror.contents[0])[2] #generates a url string for the mirror
        tmp_mirror = Mirror(url=mirror_url) #creates a mirror object for the mirror
        tmp_mirror.get_dbdata() #gets data for the mirror
        qset.add_mirror(tmp_mirror) #adds the mirror object to the qset object
     
    for mirror in qset.mirrors:
        print (mirror)
        
if __name__ == '__main__':
    main("http://www.hsquizbowl.org/db/questionsets/747")
```

## V1. First attempt

```python
#import beautifulsoup and urllib.request classes

from bs4 import BeautifulSoup as bs
import urllib.request as ulib
import pprint as pp
import re
from datetime import datetime as dt

#setup globals
DB_URL, QSET_URL = "", ""

#setting up some regex for making my life easier
delimiters = '>', '"', '<',
regexPattern = '|'.join(map(re.escape, delimiters))

#modular setup allows us to reuse the hsquizbowl.org string elsewhere.
def setup():
    global DB_URL, QSET_URL
    DB_URL = "http://www.hsquizbowl.org/db/"
    QSET_URL = "questionsets/754/"
    return

def parse_line(text):
    '''Right now this is accomplishing the task of parsing out 
    the delims listed above and splitting up the string being
    passed in.
    
    This data coming in here is very different for different cases
    so this is generic right now.'''
    return re.split(regexPattern, str(text))

def get_title_and_desc(soup):
    '''
    <DIV class="MultilineHeading">
        <H2>ACF Fall 2014: HS mirror</H2>
        <H5>High school tournament on November 08, 2014</H5>
    </DIV>
    '''
    data = soup.findAll("div", {'class':'MultilineHeading'})
    title = parse_line(data[0].contents[1])[2] #I don't know of more elegant ways to parse this yet
    tag = parse_line(data[0].contents[3])[2]
    return (title, tag)

def get_loc_and_addr(soup):
    '''
    Right the location name and address are being parsed here.
    
    <!-- TOURNAMENT DETAILS -->
                        <P><SPAN class="FieldName">Host location:</SPAN> NIU</P>
                        <P><SPAN class="FieldName">Address:</SPAN> DuSable Hall, DeKalb, Illinois 60115</P>
                ...
                
    There is a bunch of other stuff that gets loaded, too, which we don't need right now.
    
    Maybe I can talk to Dan and Jeff about putting a separate tag
    '''
    data = soup.findAll('span', {'class':'FieldName'})
    loc, addr = "", ""
    for datum in data:
        if 'Host' in datum.contents[0]:
            loc = str(datum.next_sibling)
            continue
        elif 'Address' in datum.contents[0]:
            addr = str(datum.next_sibling)
            continue
    return (loc, addr)

def get_date_from_tag(tag):
    '''
    Right now the only place to find the date for the tournament is from the end of the tagline.
    
    I think there could be a discrete field or a tag around it.
    
    Again, probably talk to Dan/Jeff about it.
    '''
    date, date_str = [], ""
    vals = tag.split(" ")
    for i in [-3,-2,-1]:
        date_str += vals[i]
    date = dt.strptime(date_str, "%B%d,%Y")
    return date

def gather_mirror_data(link):
    '''
    Given a link for the mirror returns a dict object ret_val
    Calls helper functions above as necessary.
    '''
    date = None
    soup = bs(ulib.urlopen(link).read())
    title, tag = get_title_and_desc(soup) #get title and description
    loc, addr = get_loc_and_addr(soup) #get location and address
    if tag != "":
        date = get_date_from_tag(tag) #parse out the date from the description
    else :
        date = "Unknown"
    return ({'title': title, 'date': date, 'location': loc, 'address': addr})

'''
Program starts here

Everything above is a helper function
'''
def main():
    
    setup()
    
    global DB_URL, QSET_URL
    
    mirror_urls = []
    
    soup = bs(ulib.urlopen(''.join((DB_URL, QSET_URL))).read()) #generate url, read the url, parse it using bs
    
    '''
    We are interested in the "Tournaments" linked to our question set.
    They are listed as:
    <span class="Name"><a href="tournaments/2310/">ACF Fall 2014: HS mirror</a></span>
    ... etc.
    '''
    
    mirrors_raw=soup.findAll('span',{'class':'Name'})
    for mirror in mirrors_raw:
        details = parse_line(mirror.contents[0])
        mirror_urls.append(DB_URL + details[2])
    
    mirror_data = []
    
    for link in mirror_urls:
        mirror_data.append(gather_mirror_data(link))
    
    pp.pprint(mirror_data)
    
if __name__ == '__main__':
    main()
```

```python
from app import db
from datetime import datetime as dt

class Tournament(db.Model):
    
    def _get_yr():
        return dt.today().year
    
    id = db.Column(db.Integer,
                   primary_key = True)
    
    name = db.Column(db.String(64),
                     index = True,
                     unique = True,
                     nullable=False)
    
    year = db.Column(db.Date,
                     default = _get_yr)
    
    difficulty = db.Column(db.String(64))
    
    editor = db.Column(db.Integer,
                       db.ForeignKey('User.id'))
    
    def __init__(self, name, year, difficulty, editor):
        self.name = name
        self.year = year
        self.difficulty = difficulty,
        self.editor = editor
    
    def __repr__(self):
        return '<Tournament %r>' % (self.name)

class Mirror(db.Model):
    id = db.Column(db.Integer,
                   primary_key = True)
    
    name = db.Column(db.String(200),
                     index = True
                     nullable = False)
    
    tournament = db.Column(db.Integer,
                           db.ForeignKey('Tournament.id'))
    
    host = db.Column(db.Integer,
                     db.ForeignKey('Institution.id'))
    
    td = db.Column(db.Integer,
                   db.ForeignKey('User.id'))
    
    date = db.Column(db.Date,
                     nullable = False)
    
    time = db.Column(db.Date,
                     nullable = False)
    
    address = db.Column(db.String(200),
                        nullable = False)
    
    location = db.Column(db.String(200),
                         unllable = False)
    
    def __init__(self, tournament, host, name, date, contact, address, location):
        self.tournament = tournament
        self.host = host
        self.name = name
        self.date = date
        self.contact = contact
        self.address = address
        self.location = location
    
    def __repr__(self):
        return '<Mirror %r>' % (self.name)

class Institution(db.Model):
    id = db.Column(db.Integer,
                   primary_key = True)
    
    name = db.Column(db.String(64),
                     index = True,
                     unique = True,
                     nullable=False) 
    
    address = db.Column(db.String(200),
                        nullable = False)
    
    members = db.relationship('User',
                           db.ForeignKey('User.id'),
                           backref='affiliation',
                           nullable = True)
    
    def __init__(self, name, address):
        self.name = name
        self.address = address
    
class Team(db.Model):
    id = db.Column(db.Integer,
                   primary_key = True)
    
    institution = db.Column(db.Integer,
                            db.ForeignKey('Institution.id'),
                            nullable = False)
    
    name = db.Column(db.String(50),
                     nullable = False)
    
    is_open = db.Column(db.Boolean)
    
    mirror = db.Column(db.Integer, 
                       db.ForeignKey('Mirror.id'),
                       nullable = False)
    
    def __init__(self, institution, name, is_open = False, mirror):
        self.institution = institution
        self.name = name
        self.is_open = is_open
        self.mirror = mirror
        
    
class User(db.Model):
    id = db.Column(db.Integer,
                   primary_key = True)
    
    name = db.Column(db.String(64),
                     index = True,
                     nullable = False)
    
    email = db.Column(db.String(100),
                      unique = True,
                      nullable = False)
    
    institution = db.Column(db.Integer,
                            db.ForeignKey('Institution.id'))
    
    phone = db.Column(db.String(20),
                      nullable = True)
    
    mirror = db.relationship('Mirror',
                             lazy='dynamic',
                             backref = 'td')
    
    tournament = db.relationship('Tournament',
                                 lazy = 'dynamic',
                                 backref = 'editor')
    
    def __init__(self, name, email="unknown@email.com", institution=None, phone=None):
        self.name = name
        self.email = email
        self.contact = contact
        self.institution = institution
        self.phone = phone
        
    def __repr__(self):
        return '<User %r>' % (self.name)
    

```
