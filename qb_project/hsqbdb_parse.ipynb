```python
#import beautifulsoup and urllib.request classes

from bs4 import BeautifulSoup as bs
import urllib.request as ulib
import pprint as pp
import re
from datetime import datetime as dt

#setup globals
DB_URL, QSET_URL = "", ""

#setting up some regex for making my life easier
delimiters = '>', '"', '<',
regexPattern = '|'.join(map(re.escape, delimiters))

#modular setup allows us to reuse the hsquizbowl.org string elsewhere.
def setup():
    global DB_URL, QSET_URL
    DB_URL = "http://www.hsquizbowl.org/db/"
    QSET_URL = "questionsets/754/"
    return

def parse_line(text):
    '''Right now this is accomplishing the task of parsing out 
    the delims listed above and splitting up the string being
    passed in.
    
    This data coming in here is very different for different cases
    so this is generic right now.'''
    return re.split(regexPattern, str(text))

def get_title_and_desc(soup):
    '''
    <DIV class="MultilineHeading">
        <H2>ACF Fall 2014: HS mirror</H2>
        <H5>High school tournament on November 08, 2014</H5>
    </DIV>
    '''
    data = soup.findAll("div", {'class':'MultilineHeading'})
    title = parse_line(data[0].contents[1])[2] #I don't know of more elegant ways to parse this yet
    tag = parse_line(data[0].contents[3])[2]
    return (title, tag)

def get_loc_and_addr(soup):
    '''
    Right the location name and address are being parsed here.
    
    <!-- TOURNAMENT DETAILS -->
                        <P><SPAN class="FieldName">Host location:</SPAN> NIU</P>
                        <P><SPAN class="FieldName">Address:</SPAN> DuSable Hall, DeKalb, Illinois 60115</P>
                ...
                
    There is a bunch of other stuff that gets loaded, too, which we don't need right now.
    
    Maybe I can talk to Dan and Jeff about putting a separate tag
    '''
    data = soup.findAll('span', {'class':'FieldName'})
    loc, addr = "", ""
    for datum in data:
        if 'Host' in datum.contents[0]:
            loc = str(datum.next_sibling)
            continue
        elif 'Address' in datum.contents[0]:
            addr = str(datum.next_sibling)
            continue
    return (loc, addr)

def get_date_from_tag(tag):
    '''
    Right now the only place to find the date for the tournament is from the end of the tagline.
    
    I think there could be a discrete field or a tag around it.
    
    Again, probably talk to Dan/Jeff about it.
    '''
    date, date_str = [], ""
    vals = tag.split(" ")
    for i in [-3,-2,-1]:
        date_str += vals[i]
    date = dt.strptime(date_str, "%B%d,%Y")
    return date

def gather_mirror_data(link):
    '''
    Given a link for the mirror returns a dict object ret_val
    Calls helper functions above as necessary.
    '''
    date = None
    soup = bs(ulib.urlopen(link).read())
    title, tag = get_title_and_desc(soup) #get title and description
    loc, addr = get_loc_and_addr(soup) #get location and address
    if tag != "":
        date = get_date_from_tag(tag) #parse out the date from the description
    else :
        date = "Unknown"
    return ({'title': title, 'date': date, 'location': loc, 'address': addr})

'''
Program starts here

Everything above is a helper function
'''
def main():
    
    setup()
    
    global DB_URL, QSET_URL
    
    mirror_urls = []
    
    soup = bs(ulib.urlopen(''.join((DB_URL, QSET_URL))).read()) #generate url, read the url, parse it using bs
    
    '''
    We are interested in the "Tournaments" linked to our question set.
    They are listed as:
    <span class="Name"><a href="tournaments/2310/">ACF Fall 2014: HS mirror</a></span>
    ... etc.
    '''
    
    mirrors_raw=soup.findAll('span',{'class':'Name'})
    for mirror in mirrors_raw:
        details = parse_line(mirror.contents[0])
        mirror_urls.append(DB_URL + details[2])
    
    mirror_data = []
    
    for link in mirror_urls:
        mirror_data.append(gather_mirror_data(link))
    
    pp.pprint(mirror_data)
    
if __name__ == '__main__':
    main()
```
